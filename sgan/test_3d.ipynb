```python
import infer
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


lat_min, lat_field, long_min, long_field, alt_min, alt_field = 3.1933607061071148,50.80615587860828,78.42022293891259,56.574887353137626,0.0,43734.576275625324

# lat_min, lat_field, long_min, long_field, alt_min, alt_field = 3.1933607061071148,50.80410785495143,78.42022293891259,56.574887353137626,10599.32828794153,33135.2479876838
```

```python
def show_t(obs_traj, pred_traj_fake, pred_traj_gt):
    
    obs_traj[:, :, 0] = obs_traj[:, : ,0] * lat_field + lat_min
    pred_traj_fake[:, :, 0] = pred_traj_fake[:, : ,0] * lat_field + lat_min
    pred_traj_gt[:, :, 0] = pred_traj_gt[:, : ,0] * lat_field + lat_min
    
    obs_traj[:, :, 1] = obs_traj[:, : ,1] * long_field + long_min
    pred_traj_fake[:, :, 1] = pred_traj_fake[:, : ,1] * long_field + long_min
    pred_traj_gt[:, :, 1] = pred_traj_gt[:, : ,1] * long_field + long_min
   
    obs_traj[:, :, 2] = obs_traj[:, : , 2] * alt_field + alt_min
    pred_traj_fake[:, :, 2] = pred_traj_fake[:, : , 2] * alt_field + alt_min
    pred_traj_gt[:, :, 2] = pred_traj_gt[:, : , 2] * alt_field + alt_min
   
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    
    for i in range(1):
        ax.plot(obs_traj[:,i,0], obs_traj[:,i,1], obs_traj[:,i,2], color='r', label='input', linewidth=5)
        ax.plot(pred_traj_fake[:,i,0], pred_traj_fake[:,i,1], pred_traj_fake[:,i,2], color='b', label='prediction', linewidth=5)
        ax.plot(pred_traj_gt[:,i,0], pred_traj_gt[:,i,1], pred_traj_gt[:,i,2], color='y', label='groudtruth', linewidth=5)
    plt.show()
    # plt.legend()
```

```python
path = 'exp13_with_model.pt'
infer1 = infer.Infer(use_cuda=0)
infer1.load_model(path)
```
