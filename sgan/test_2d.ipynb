```python
import infer
import matplotlib.pyplot as plt
```

```python
def show_t(obs_traj, pred_traj_fake, pred_traj_gt):
    for i in range(1):
        plt.plot(obs_traj[:,i,0], obs_traj[:,i,1], color='r', label='input', linewidth=5)
        plt.plot(pred_traj_fake[:,i,0], pred_traj_fake[:,i,1], color='b', label='prediction', linewidth=5)
        plt.plot(pred_traj_gt[:,i,0], pred_traj_gt[:,i,1], color='y', label='groudtruth', linewidth=5)
    plt.legend()
```

```python
path = 'test_china_with_model.pt'
infer1 = infer.Infer(use_cuda=0)
infer1.load_model(path)
```

```python
obs_traj, pred_traj_fake, pred_traj_gt = infer1.infer()
show_t(obs_traj, pred_traj_fake, pred_traj_gt)
```

```python
obs_traj, pred_traj_fake, pred_traj_gt = infer1.infer()
show_t(obs_traj, pred_traj_fake, pred_traj_gt)
```
